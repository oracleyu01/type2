```python
# ▣ 빅데이터 분석 기사 실기 유형2번 코드 암기하기(분류 파트)

# ■ 시험환경 만들기1. sklearn 에서 유방암 환자 데이터 불러오기

from sklearn.datasets import load_breast_cancer
brst = load_breast_cancer()
x, y = load_breast_cancer(return_X_y=True)
```

```python
# ■ 시험환경 만들기2. 학습 데이터와 정답 데이터 만들기

import pandas as pd

col = brst.feature_names                 # 컬럼명 불러오기
X = pd.DataFrame(x , columns=col)        # 학습 데이터
y = pd.DataFrame(y, columns=['cancer'])  # 정답 데이터
X
```

```python
# ■ 시험환경 만들기3. 훈련 데이터와 테스트 데이터를 생성합니다.

# 훈련 데이터와 테스트 데이터를 분리합니다.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test  = train_test_split(X, y,test_size=0.2, random_state=1)

# 만든 데이터를 시험환경에 저장합니다.
x_train.to_csv("X_train.csv", index=False)
y_train.to_csv("y_train.csv", index=False)
x_test.to_csv("X_test.csv", index=False)
```

```python
# ■ 시험문제 풀기 시작

# 1. 기계 학습 시킬 데이터를 불러옵니다.
import pandas as pd
X_test = pd.read_csv("X_test.csv")
X_train = pd.read_csv("X_train.csv")
y_train = pd.read_csv("y_train.csv")
```

```python
# 2. 데이터를 정규화 합니다.

# 훈련 데이터 정규화
from  sklearn.preprocessing  import MinMaxScaler

scaler=MinMaxScaler()

scaler.fit(x_train)
x_train2 = scaler.transform(x_train)

# 테스트데이터 정규화
from  sklearn.preprocessing  import MinMaxScaler

scaler=MinMaxScaler()

scaler.fit(x_test)
x_test2 = scaler.transform(x_test)
```

```python
# 3. 정답 데이터 만들기

y_train = pd.read_csv("y_train.csv")
y = y_train['cancer']
y
```

```python
# 4. 모델 생성

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
```

```python
# 5. 모델훈련
model.fit(x_train2, y)
```

```python
# 6. 모델예측
result = model.predict(x_test2)
result
```
